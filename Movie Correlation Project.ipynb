```python
# Using factorize - this assigns a random numeric value for each unique categorical value

df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

```python
for col_name in df.columns:
    if(df[col_name].dtype == 'object'):
        df[col_name]= df[col_name].astype('category')
        df[col_name] = df[col_name].cat.codes
```

```python
# Step 1: Identify categorical columns in the DataFrame
cat_columns = df.select_dtypes(include=['category', 'object']).columns

# Step 2: Convert categorical columns to category codes
df[cat_columns] = df[cat_columns].apply(lambda x: x.astype('category').cat.codes)

# Display the updated DataFrame
df
```

```python
# Handle infinite values
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop rows with NaN values in 'rating' and 'gross'
df = df.dropna(subset=['rating', 'gross'])

sns.stripplot(x="rating", y="gross", data=df)

plt.title("Gross Revenue by Rating")
plt.show()
```
